**Multiple ways of creating a figure:** ( Using Line chart as an example)

```python
import matplotlib.pyplot as plt
#1. Use the plot interface directly  - This approach internally calls figure and axes objects via the plot class 
plt.plot( [ 1,2,3],[1,2,3])
plt.title('Call Plot directly')

#2. Explicitly Create a figure and then get the axes of the figure to plot the chart 
fig = plt.figure()
ax = fig.gca()
ax.plot( [ 1,2,3],[1,2,3])
ax.set_title('Call Axes object')

#3.Create subplots and call the individual axes within the subplot figures
fig1, ax1 = plt.subplots(2, 2, sharex = True, sharey = True) # can also use nrows = 2, ncols = 2, # if there are few subplots one of the return arugments can be tuple e.g. ( ax0 , ax1) instead of an axes array
ax1[0,0].plot( [ 1,2,3],[1,2,3])
ax1[0,0].set_title('Call Axes object within a Subplot')

plt.show()
```

**Scatter plot, bar chart, Histogram**

```python
import matplotlib.pyplot as plt 
import numpy as np
np.random.seed(111)

x = np.random.randint(1, 10, 10)
y = 5+np.random.randint(1, 10, 10)
fig, (ax1,ax2,ax3) = plt.subplots(1,3, figsize = (20,5))
# ax1 = scatter 
ax1.scatter(x,y,  c='r', edgecolor = 'b')
ax1.set_title('Test Scatter')
ax1.set_xlabel('x axis')
ax1.set_ylabel( 'y axis')

#ax2 = bar chart 
ax2.bar(x,height = y, color = 'purple', label = 'random')
ax2.legend()

#ax3 = histogram 
ax3.hist([x,y])  ## stacked histogram 

plt.show()
```

**Work with Splines**


```python
import matplotlib.pyplot as plt 
import numpy as np
np.random.seed(111)

x = np.random.randint(1, 10, 10)
y = 5+np.random.randint(1, 10, 10)
fig, ax = plt.subplots()

ax.bar(x,height = y, color =(1.,0.,0.) )

# ax.bar(x,height = y, color ='r' )

ax.spines['left'].set_color('green')
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['left'].set_linewidth(5)


plt.show()
```
